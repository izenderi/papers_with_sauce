```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("pc.csv")

# Font size
plt.rcParams.update({'font.size': 18})

# Reverse the order of the x-axis values to maintain a descending order
df = df.sort_values(by=df.columns[0], ascending=True)  # Sort in ascending order first

# Create a figure and two subplots sharing the x-axis
fig, ax1 = plt.subplots(figsize=(8, 6))

# Create a second y-axis
ax2 = ax1.twinx()

# Plot pose drop on ax1
ax1.plot(df.iloc[:, 0], df.iloc[:, 1], marker='o', linestyle='-', label=df.columns[1], color='C0')
ax1.plot(df.iloc[:, 0], df.iloc[:, 2], marker='x', linestyle='-', label=df.columns[2], color='C1')

# Plot edge similarity on ax2
ax2.plot(df.iloc[:, 0], df.iloc[:, 3], marker='o', linestyle='--', label=df.columns[3], color='C0')
ax2.plot(df.iloc[:, 0], df.iloc[:, 4], marker='x', linestyle='--', label=df.columns[4], color='C1')

# Formatting
# ax1.set_xlabel(df.columns[0])
# ax1.set_ylabel("Pose Drop (%)", color='b')
# ax2.set_ylabel("Edge Similarity (%)", color='r')

# Reverse the x-axis direction to maintain large-to-small order
ax1.invert_xaxis()

# Grid and legend only ax1 x axis grid
ax1.grid(True, axis='x')

# ax1.legend(loc='upper left')
# ax2.legend(loc='upper right')

ax1.set_ylim(0, 100)
ax2.set_ylim(0, 1)

# Save and show the plot
plt.savefig("pc.png", dpi=300)
plt.show()
```
